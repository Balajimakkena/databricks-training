### 1.Create bonus column as 10% of salary.


```python
from pyspark.sql.functions import col

df1 = df.withColumn(
    "bonus",
    col("salary") * 0.10
).display()
```

### 2.Create annual_salary column.


```python
df1= df.withColumn(
    "annual_salary",
    col("salary") * 12
).display()
```

### 3.Create tax column as 5% of salary.


```python
df1 = df.withColumn(
    "tax_column",
    col("salary") * 0.05
).display()
```

### 4.Create updated_salary by adding 5000.


```python
df1 = df.withColumn(
                    "Updated_salary",
                    col("salary") + 50000
).display()
```

### 5.Create salary_category column:
### High
### Medium
### Low


```python
from pyspark.sql.functions import when

df1 = df.withColumn(
    "salary_category",
    when(col("salary") >= 80000, "High")
    .when(col("salary") >= 50000, "Medium")
    .otherwise("Low")
).display()
```

### 6.Create age_group column:
### Young
### Adult


```python
df1 = df.withColumn(
    "age_group",
    when(col("age")>25,"Adult")
    .when(col("age")<=25,"Young")).display()
```

### 7.Create location column by combining city and department.


```python
from pyspark.sql.functions import concat_ws
df1 = df.withColumn(
                    "loaction",
                    concat_ws("-",col("city"),col("department"))
).display()
```

### 8.Create increment_salary column with 15% hike.


```python
df1 = df.withColumn(
    "increment_salary",
    col("salary") * 1.15
).display()
```

### 9.Create experience_status column based on joining year.


```python
from pyspark.sql.functions import col, when, year

df = df.withColumn(
    "experience_status",
    when(year(col("joining_date")) < 2020, "Experienced")
    .otherwise("Fresher")
)

df.display()
```

### 10.Create name_length column using employee name.


```python
from pyspark.sql.functions import length

df = df.withColumn(
    "name_length",
    length(col("emp_name"))
).display()
```

### 11.Create is_high_salary column using condition.


```python
df1 = df.withColumn(
    "is_high_salary",
    when(col("salary") > 70000, True)
    .otherwise(False)
).display()
```

### 12.Create joining_year column from joining_date.


```python
from pyspark.sql.functions import year

df = df.withColumn(
    "joining_year",
    year(col("joining_date"))
).display()
```

### 13.Create salary_after_tax column.


```python
df1 = df.withColumn(
    "salary_after_tax",
    col("salary") - (col("salary") * 0.05)
).display()
```

### 14.Create department_code column.


```python
df1 = df.withColumn(
    "department_code",
    when(col("department") == "IT", "IT01")
    .when(col("department") == "HR", "HR01")
    .otherwise("GEN01")
).display()
```

### 15.Create double_salary column

```python
df1 = df.withColumn(
    "double_salary",
    col("salary") * 2
).display()
```
