### 1.Filter employees whose salary is greater than 70000.


```python
from pyspark.sql.functions import col,lit
df.filter(
    col("salary") > 70000
).select(col("emp_name"), col("salary")).display()
```

### 2.Filter employees from Hyderabad.


```python
df.filter(
    col("city")=="Hyderabad"
).select(col("emp_name"), col("city")).display()
```

### 3.Filter employees whose age is less than 25.


```python
df.filter(
    col("age")<25
).select(col("emp_name"), col("age")).display()
```

### 4.Filter employees from IT department.


```python
df.filter(
    col("department")== "IT"
).select(col("emp_name"), col("department")).display()
```

### 5.Filter employees whose designation is Developer.


```python
df.filter(
    col("designation")== "Developer"
).select(col("emp_name"), col("designation")).display()
```

### 6.Filter employees whose salary is between 50000 and 80000.


```python
df.filter(
    col("salary").between( 50000 , 80000)
).select(col("emp_name"), col("salary")).display()
```

### 7.Filter employees whose city is Bangalore.


```python
df.filter(
    col("city")=="Bangalore"
).select(col("emp_name"), col("city")).display()
```

### 8.Filter employees who joined after 2022-01-01.


```python
df.filter(
    col("joining_date")>=lit("2022-01-01")
).select(col("emp_name"), col("joining_date")).display()
```

```python
df.filter(
    col("age")> 30
).select(col("emp_name"), col("age")).display()
```
