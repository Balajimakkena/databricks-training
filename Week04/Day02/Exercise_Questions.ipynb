### withColumn() Transformation Questions
 


### 1.Add a new column bonus with 10% of salary.


```python
df = emp_df.withColumn("bonus",emp_df.salary*0.1).select("name","bonus")
df.show()
```

### 2.Add a new column tax with 5% deduction.


```python
df = emp_df.withColumn("tax",emp_df.salary*0.05).select("name","tax")
df.show()
```

```python
df = emp_df.withColumn("salary_increment",emp_df.salary+5000).select("name","salary_increment")
df.show()
```

### 4.Create a new column age_group based on age.
![](path)

```python
emp_df.withColumn("age_group",
                  when(emp_df.age < 18,"Young")
                  .when((emp_df.age >=18) & (emp_df.age <=50),"adult")
                  .otherwise("senior")).show()
```

### agg() Questions
 


## ### 1.Find total salary of all employees.


```python
emp_df.select("salary").agg({"salary": "sum"}).show()
```

### 2.Find average employee age.


```python
emp_df.select("age").agg({"age": "avg"}).show()
```

### 3.Find max and min salary together.


```python
emp_df.agg(
    max("salary").alias("max_salary"), 
    min("salary").alias("min_salary")).show()
```

### 4.Find count of employees.


```python
emp_df.agg(
    count("emp_id").alias("emp_count")
).show()
```

### 5.Find average salary and average age department-wise.

```python
emp_df.groupBy("department").agg(
    avg("salary").alias("Avg_salary"),
    avg("age").alias("avg_age")
).show()
```
