### union() Questions

```python
new_emp_data = [
(111, "Teja", "Developer", "IT", 72000, "Hyderabad", 26, "2023-01-10",
"Male"),
(112, "Divya", "Analyst", "Analytics", 65000, "Bangalore", 25,
"2022-11-05", "Female")
]
new_emp_df = spark.createDataFrame(new_emp_data, columns)
```

 
### 1.Union emp_df with new_emp_df.


```python
emp_df.union(new_emp_df).show()
```

### 2.Count total rows after union.
![](path)

```python
df_1=emp_df.union(new_emp_df)
df_1 .count()
```

### 3.Remove duplicates after union.


```python
df_1=emp_df.union(new_emp_df).distinct()
df_1 .show()
```

### 4.Find all IT employees after union.


```python
df_1.filter(df_1.department=="IT")\
    .select("name","department").show()
```

### 5.Sort unioned dataframe by salary.

```python
df_1.orderBy("salary",ascending=False)\
    .select("name","salary").show()
```

### unionByName() Questions

 
### 1.Create another dataframe with shuffled columns and perform unionByName.


```python
emp_df_shuffled = emp_df.select(
    "name",
    "emp_id",
    "department",
    "salary",
    "designation",
    "city",
    "age",
    "joining_date",
    "gender"
)

emp_df_shuffled.show()
result_df = emp_df.unionByName(emp_df_shuffled)

result_df.show()
```

### 2.Use allowMissingColumns=True.


```python
from pyspark.sql.functions import lit

emp_df_new = emp_df.withColumn("bonus", lit(5000))

emp_df.unionByName(
    emp_df_new,
    allowMissingColumns=True
).show()
```

### 3.Add extra column and perform unionByName.


```python
from pyspark.sql.functions import lit

emp_df_new = emp_df.withColumn("bonus", lit(5000))

emp_df.unionByName(
    emp_df_new,
    allowMissingColumns=True
).show()
```

### 4.Compare union vs unionByName.


```python
emp_df_shuffled = emp_df.select(
    "name",
    "emp_id",
    "department",
    "salary",
    "designation",
    "city",
    "age",
    "joining_date",
    "gender"
)
emp_df.union(emp_df_shuffled).show()
emp_df.unionByName(emp_df_shuffled).show()
```
